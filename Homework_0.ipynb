```python
def colour_to_gray(img): ##convert to gray
    
    im = plt.imread(img)
    im_r = im[:,:,0]
    im_g = im[:,:,1]
    im_b = im[:,:,2]
    gray_image = 0.2989 * im_r + 0.5870 * im_g + 0.1140 * im_b
    
    return gray_image

plt.imshow(colour_to_gray("bg1.jpg"))
```

```python
while True:
    # `success` is a boolean and `frame` contains the next video frame
    success, frame = video_cap.read()
    cv2.imshow("frame", frame)
    # wait 20 milliseconds between frames and break the loop if the `q` key is pressed
    if cv2.waitKey(20) == ord('q'):
        break

# we also need to close the video and destroy all Windows
video_cap.release()
cv2.destroyAllWindows()
```
